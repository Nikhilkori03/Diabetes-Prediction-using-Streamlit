# Separating the Data and Label

```python
X = DC.drop(columns = 'Outcome', axis=1)
Y = DC['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

# Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardization_data = scaler.transform(X)
print(standardization_data)
```

```python
X = standardization_data
Y = DP.Outcome
```

```python
print(X)
print(Y)
```

# Training Data and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Traning the Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

Training the support vector Machine Classifier

```python
classifier.fit(X_train, Y_train)
import pickle

file = "Diabetes_model.sav"
pickle.dump(classifier,open(file,"wb"))
```

```python
pickle.load(open("Diabetes_model.sav",'rb'))
```

# Making a Predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

#changing the input data to numpy array
input_data_as_numpy_arrary = np.asarray(input_data)

#re-shape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_arrary.reshape(1, -1)

#standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('The person is not diabetic')
else :
    print('The person is diabetic')
```
